```python
from matplotlib import pyplot as plt
import csv

def initialize_data(file_name: str):
    # Load csv file
    rows = []
    with open(f'out/{file_name}', 'r') as csv_file:
        csv_reader = csv.reader(csv_file)

        _ = next(csv_reader)

        for row in csv_reader:
            rows.append(row)

        print(f"{csv_reader.line_num - 1} rows appended")
    return rows

results = initialize_data('analysis.csv')
crowd_sourced_performance = 0.8557

names = [row[0] for row in results]
names.append('crowd-sourced')

acc = [float(row[-1]) for row in results]
acc.append(crowd_sourced_performance)

plt.bar(names, acc, color='#ff00fe', label='best synonym guess')
plt.style.use('dark_background')

plt.title('Performance per model')

plt.xlabel("Models")
plt.ylabel("Accuracy")
#plt.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.legend(loc='lower right')

plt.savefig('out/performance_plot.pdf', bbox_inches='tight')
```
